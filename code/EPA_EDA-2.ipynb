```python
import pandas as pd

# Data as provided
data = {
    "Fiscal Year": ["FY 2023", "FY 2022", "FY 2021", "FY 2020", "FY 2019", "FY 2018", "FY 2017", "FY 2016",
                    "FY 2015", "FY 2014", "FY 2013", "FY 2012", "FY 2011", "FY 2010", "FY 2009", "FY 2008"],
    "Enacted Budget": [10135433000, 9559485000, 9237153000, 9057401000, 8849488000, 8824488000, 8058488000,
                       8139887000, 8139887000, 8200000000, 7901104000, 8449385000, 8682117000, 10297864000,
                       7643674000, 7472324000]
}

# Create DataFrame
epa_df = pd.DataFrame(data)
```

```python
import matplotlib.pyplot as plt

# Convert 'Fiscal Year' to a more sortable format (e.g., from 'FY 2008' to 2008)
epa_df['Year'] = epa_df['Fiscal Year'].str.split(' ').str[1].astype(int)

# Sort DataFrame by 'Year'
epa_df_sorted = epa_df.sort_values(by='Year')

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(epa_df_sorted['Year'], epa_df_sorted['Enacted Budget'] / 1e9, marker='o', linestyle='-', color='blue')
plt.title('EPA Budget from 2008 to 2023')
plt.xlabel('Year')
plt.ylabel('Enacted Budget (in billions)')
plt.grid(True)
plt.xticks(epa_df_sorted['Year'], rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Adding the proposed budget for 2024 to the existing sorted DataFrame
proposed_budget_2024 = 12.083e9  # 12.083 billion dollars
df_proposed = epa_df_sorted.append({'Year': 2024, 'Enacted Budget': proposed_budget_2024}, ignore_index=True)

# Plotting with a dotted line from 2023 to 2024 for the proposed budget
plt.figure(figsize=(10, 6))
# Actual budgets up to 2023
plt.plot(epa_df_sorted['Year'], epa_df_sorted['Enacted Budget'] / 1e9, marker='o', linestyle='-', color='blue', label='Actual Budget')
# Proposed budget for 2024 with a dotted line
plt.plot(df_proposed['Year'].tail(2), df_proposed['Enacted Budget'].tail(2) / 1e9, marker='o', linestyle=':', color='blue', label='Proposed Budget for 2024')
plt.title('EPA Budget from 2008 to 2024 (with Proposed Budget)')
plt.xlabel('Year')
plt.ylabel('Enacted Budget (in billions)')
plt.grid(True)
plt.xticks(df_proposed['Year'], rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
